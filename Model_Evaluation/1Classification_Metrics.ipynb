```python
# now we are going to talk about 
# ! precision 

# suppose we had set a covid camp and in that camp 10 people came and among those 10 6 people were actually sick so the precision of model becomes 
# 6/10 = 60% 

# In machine learning, precision measures the quality of a model's positive predictions 
# psostive prediction means the exact correct answer whcih we were expecting at that time 

#   !  Precision = TP/(TP+FP)
#     where TP is True Positives and FP is False Positives
```

```python
# Now we are going to learn about the 
#! Recall 

# Recall is the number of cases predicted by our model among those predicted what were actual positive cases 

#! suppose 8 people had covid but our machine predicted that only 6 have covid so recall is 6/8 = 75%
```

```python

#! F1

# Now F1 is a balance between the precision and Recall 
#! suppose there were 100 people and our model said 5% people are sick and rest 95% are healthy so here we can see that our model is being baised so that is something we don't want, we want our model to be unbaised

#? The F1 score is a machine learning evaluation metric that measures a classification model's accuracy by calculating the harmonic mean of precision and recall. It provides a balanced performance score, especially useful for imbalanced datasets where the positive class is rare, ranging from 0 (worst) to 1 (best).

```
