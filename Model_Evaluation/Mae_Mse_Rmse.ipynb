```python

#! 1Ô∏è‚É£ MAE (Mean Absolute Error)

# Matlab:

# Pehle difference lo

# Absolute value lo (minus hata do)

# Average nikal lo

# ?Example:

# Errors = [-2, 2, -3]
# Absolute = [2, 2, 3]

# MAE=(2+2+3)/3=7/3=2.33

# üëâ Matlab:

# On average model 2.33 units galat hai.
```

```python

#! 2Ô∏è‚É£ MSE (Mean Squared Error)

# Matlab:

# Error lo

# Square karo

# Average nikal lo

#? Example:

# Errors = [-2, 2, -3]
# Square = [4, 4, 9]

# MSE=(4+4+9)/3=17/3=5.67

# 5.67
# !üî• MSE ki khas baat:

# Large errors ko jyada punishment deta hai

# Outliers se heavily affect hota hai

# Kyuki square karne se:
# 3¬≤ = 9
# 10¬≤ = 100 üò≥
```
