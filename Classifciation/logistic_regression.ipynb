```python
from sklearn.linear_model import LogisticRegression

X = [[1], [2], [3], [4], [5]] #Number of hours student studied
y = [0, 0, 1, 1, 1] #Result if the student will pass or fail like if he studied for one hour he will fail but if he studied for 2 hours he will pass

model =LogisticRegression()
model.fit(X,y)

hours = float(input("Enter how many hours you studied = ")) 

result = model.predict([[hours]])[0]

if result == 1 :
    print(f"Based on hours {hours}, you are likely to PASS")
else:
    print(f"Based on hours{hours}, you are likely to FAIL")
```
