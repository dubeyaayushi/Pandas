```python
# here we are gonna make a model to predict th efruit based on it's weight and size
# weight is in grams and size is in cm

from sklearn.neighbors import KNeighborsClassifier

X = [
    [180, 7],
    [200, 7.5], #weight and size
    [250, 8],
    [300, 8.5],
    [330, 9],
    [360, 9.5],
    [550, 7.8],
    [800, 8]
    
]

y = [0,0,0,1,1,1,2,2] #0 = apple 1 = orange

model = KNeighborsClassifier(n_neighbors=3)

model.fit(X,y)

weight = float(input("Enter the weight in grams:"))
size =  float(input("Enter the size in cm: "))

prediction = model.predict([[weight, size]])[0]
model.predict([[weight, size]])[0] means:

# ?Take this fruit's weight & size → compare with known fruits → find 3 nearest → vote → return the first prediction value.


if prediction == 0:
    print("this is likely an Apple")
elif prediction == 1 :
    print("this is orange")    
else:
    print("This is Likely mango")
    
    
    
#!     Could we use numbers other than 0 and 1?

# Yes — absolutely.

#! This would also work:

# y = [10,10,10,20,20,20]


# or

# y = [1,1,1,2,2,2]


# ?The model just learns class separation, not the meaning of numbers.

# We usually use 0 and 1 because:

# simple

# standard practice

# easy to interpret
```
