```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder


df = pd.read_csv("sample_data.csv", encoding='latin1')
# print(df) 
df.columns = df.columns.str.strip()

print(df.columns)

df_label = df.copy() #ek copy bana lo original data ki jisse humara original data kharb na ho aur abb iss copied data pe kam ya phir changes karenge 

# LabelEncoder is a scikit-learn preprocessing utility that converts categorical text data into numerical labels (integers ranging from \(0\) to \(n\_classes-1\)). It is primarily used to transform target labels (\(y\)) into a machine-readable format, such as converting ['cold', 'hot', 'warm'] to [0, 2, 1]. It is ideal for ordinal data but can introduce false numerical relationships if used on nominal data.Â 

le = LabelEncoder()
df_label['Gender_Encoded'] = le.fit_transform(df_label['Gender'])
df_label['Passed_Encoded'] = le.fit_transform(df_label['Passed'])

print('\n Label Encoded Data')
print(df_label[['1 Name' , 'Gender' , 'Gender_Encoded', 'Passed', 'Passed_Encoded']].head())
```
