```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split

data = {
    'StudyHours':[1,2,3,4,5],
    'TestScore':[40,50,60,70,80]
}

df = pd.DataFrame(data)

#standardScaler
scaler = StandardScaler()
standar_scaler= scaler.fit_transform(df)

print("Strandard Sclaer Output")
print(pd.DataFrame(standar_scaler, columns=['StudyHours','TestScore']))

scaled = MinMaxScaler()
minmaxScaled = scaled.fit_transform(df)

print("Min Max Scaled data")
print(pd.DataFrame(minmaxScaled, columns=["StudyHours","TestScore"])) 


# ! Training and Testing Data
x = df[['StudyHours']] #this is our input data 
y = df[['TestScore']] #y humlogo ka label hai matlab hai output jo humein predict karna hai 

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print('Training Data')
print(x_train)

print("Test Data")
print(x_test)
```
