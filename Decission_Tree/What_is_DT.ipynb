```python

# !this is a model which helps us to make the predictions by asking a sseries of yes and no questions based on data features

#A decision tree is a flowchart-like structure used in both data analysis and machine learning to map out a series of decisions and their potential consequences. It serves as a visual decision support tool and a predictive model, breaking down complex problems into a sequence of simple, understandable choices
```

```python
from sklearn.tree import DecisionTreeClassifier

X = [
    [7,2],
    [8,3],
    [9,8],
    [10,9],
    [11,2],
    [12,5],
    [2,3]
]

Y = [0 , 0 ,0 ,1, 2, 2,2 ]

model = DecisionTreeClassifier()

model.fit(X,Y)

size = float(input("ENter the weight in grams"))
shade = float(input("ENter the colour shade "))

result = model.predict([[size, shade]])[0]

if result == 0 :
    print("this is likely to be an apple")
elif result ==  1 :
    print("This is likely to be a Orange")
else :
    print("This is cherry")
    
```
