```python

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error
import numpy as np

data = pd.read_csv("StudentPerformanceFactors.csv")

X = data[['Hours']]
y = data[['Scores']]

model = LinearRegression()
model.fit(X,y)

predicted_score = model.predict(X)

#evaluate 
mae = mean_absolute_error(y, predicted_score)
mse = mean_squared_error(y, predicted_score)
rmse = np.sqrt(mse)

print("Mean Absolute ERROR(MAE)", mae)
print("Mean Squared ERROR(MAE)", mse)
print(" Root MEan Squared Error RMSE ERROR(MAE)", rmse)

new_prediction = float(input("enter a hour = "))
new_pred = model.predict([[new_prediction]])
print(f"Prediction for {new_prediction} is score = {new_pred}")
```
