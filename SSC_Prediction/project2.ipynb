```python
import pandas as pd
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("StudentPerformanceFactors.csv")

X = data[['Hours']]
y = data['Scores']

model = LinearRegression()
model.fit(X,y)

predicted_scores = model.predict(X)

mae = mean_absolute_error(y, predicted_scores)
mse = mean_squared_error(y, predicted_scores)
rmse = np.sqrt(mse)
r2 = r2_score(y, predicted_scores)


print("Mean Absolute ERROR(MAE)", round(mae,2))
print("Mean Squared ERROR(MAE)", round(mse,2))
print(" Root MEan Squared Error RMSE ERROR(MAE)", round(rmse,2))
print("R^2 Score(Model Accuracy): ", round(r2,4))

#histogram
plt.figure(figsize=(10,6)) #defining the plot size
plt.hist(data["Scores"], bins=10, color='pink' , edgecolor='purple')

# !ðŸ§  Bins Kya Hota Hai?

# Bins = Data ko kitne groups (intervals) me divide karna hai

# Socho tumhare Final_Score 0 se 100 ke beech hain.

# Agar:

# bins = 5

#! To histogram aise divide karega:

# 0â€“20

# 20â€“40

# 40â€“60

# 60â€“80

# 80â€“100

# 5 bars banenge.

plt.title("Distribution of FINAL EXAM SCORE")
plt.xlabel("final Exam Score")
plt.ylabel("Number of students")
plt.grid(True)
plt.show()



```
