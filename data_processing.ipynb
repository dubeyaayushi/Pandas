```python
import pandas as pd 
data={
    'Name':['Pavan','Pravnav','Swapnil','Ishan','Arjun','Om'],
    'Age':[25,None,22,23,None,26],
    'Salary':[5000,6000,7000,8000,9000,5500]    
}
df= pd.DataFrame(data)
print("Original Data")
print(df)
#to find the null values in a dataset
print(df.isnull().sum())

#droping or removing those values which have null value 
df_drop = df.dropna()
print(df_drop)

#filling the values at the null columns or rows
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print(df)


#to find out how much data is missing in the dataset
df.isnull().mean()*100
```
