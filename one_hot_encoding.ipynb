```python

# ! one hot encoding 
# jo bhi category hai usko multiple columns mein convert kar dega and each will be representing one possible value.
 
# supoose ek data hai usmein city hai abb city mein Delhi, Mumbai, Chennai, Pune hai 
# so abb iske column aise baan jayega 
# City Delhi 
# City Mumbai 
# City Chennai
# City Pune 

# so abb agr koi Delhi se hai to uska baan jayega [1,0,0,0]
# ! ye vaha apply hoti hai jaha ek column mein multiple different values hoti hai 
# and category mein sirf kuch specific category hoti hai jaise gender mein male and female

# ! What One-Hot Encoding does

# Instead of assigning numbers like Label Encoding, One-Hot Encoding creates a new column for each category.

# Each row gets:

# 1 for its category

# 0 for the others

# Example — Gender column

# Original:

# Gender
# Male
# Female
# Male
# Female

# After One-Hot Encoding:

# Gender	Female	Male
# Male	0	1
# Female	1	0
# Male	0	1
# Female	1	0

# See the difference?

# We didn’t assign numbers like:

# Male = 1
# Female = 0


# Instead, we created separate columns.

# That’s why it’s called One-Hot — only one column is “hot” (1) per row.
```
